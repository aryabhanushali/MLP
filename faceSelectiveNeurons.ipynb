The goal is to determine if face selective neurons in toponets are necessary and sufficient for discriminating faces.



-Extracting activations from the model to see which neurons respond to faces.
-Using PCA to identify the most important neurons.
-Testing necessity by removing these neurons and checking if face recognition accuracy drops.
-Testing sufficiency by using only these neurons to see if they alone can recognize faces.
-If removing them disrupts recognition, they are necessary. If using only them is enough, they are sufficient.

## **Imports**

```python
import torch
import numpy as np
from datasets import load_dataset
from torchvision import transforms
from PIL import Image
import toponets
from sklearn.decomposition import PCA
```

# DataSet+Preprocessing

```python


# Load CelebA dataset
dataset = load_dataset("flwrlabs/celeba")

# Define ImageNet preprocessing
transform = transforms.Compose([
    transforms.Resize((224, 224)),  #resize for ResNet input
    transforms.ToTensor(),  #convert to PyTorch tensor
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  #imagenet normalization
])




```

# Load Pretrained TopoNets ResNet18 Model

```python
# Load pre-trained TopoNets ResNet18 model
checkpoint_path = "resnet18_tau_10.pt"
topo_resnet18 = toponets.resnet18(tau=10.0)
checkpoint = torch.load(checkpoint_path, map_location=torch.device("cpu"))
topo_resnet18.load_state_dict(checkpoint)  # Directly load checkpoint if it's a state_dict

topo_resnet18.eval()
```

# Register a Hook to Capture Activations

```python
#registers a PyTorch hook to capture the activations of a specific layer in the topo_resnet18 model.
activation = {} #store activiations

def hook_fn(module, input, output):

    activation["face_neurons"] = output.detach() #output detach prevents gradient tracking
#layer 4 is where hook is registered
# gets executed when a forward pass reaches the specified layer.
hook = topo_resnet18.layer4.register_forward_hook(hook_fn)

```

# Process Multiple Images & Collect Activations

```python
# Process multiple images and collect activations
num_images = 300  # Ensure enough samples for PCA
activations_list = [] #store activations of each image

for i in range(num_images):
    image = dataset["train"][i]["image"]  #load image
    if not isinstance(image, Image.Image):
        image = Image.fromarray(image)  #convert NumPy array to PIL if needed
    input_tensor = transform(image).unsqueeze(0)  #Applies preprocessing

    with torch.no_grad():
        _ = topo_resnet18(input_tensor)  # Run inference
        #Feeds the image into the model, triggering the forward hook

    activations_list.append(activation["face_neurons"].numpy().flatten())  # Store activations

hook.remove()  # Remove hook after collecting activations

# Convert activations list to NumPy array
flattened_activations = np.array(activations_list)

print("Activations Shape for PCA:", flattened_activations.shape)  # num_images, feature_size
```

# PCA

```python
# Ensure PCA has at least 2 samples


num_samples, num_features = flattened_activations.shape
if num_samples < 2:
    raise ValueError(f"PCA requires at least 2 samples, but got {num_samples}.")

# Set n_components safely
#reduces the dataset to a lower-dimensional representation.
#prevent errors by taking min of 20(upper limit), num_samples, num_features so you dont req more than available
n_components = min(20, num_samples, num_features)


# Apply PCA
pca = PCA(n_components=n_components) #create pc model with n comp
principal_components = pca.fit_transform(flattened_activations) #Fits PCA on the high-dimensional flattened_activations
#Transforms the activations into a lower-dimensional space (300 × n_components).

# Print explained variance ratio
print(f"Using {n_components} principal components.")
print("Explained Variance Ratios:", pca.explained_variance_ratio_)
#explained variance ratio tells us how much information each principal component captures





#output
#with 20 PCA: PC1 captured 4.86%, PC2 captured 3.16%...
#adding up all 20 = 32.9


```
